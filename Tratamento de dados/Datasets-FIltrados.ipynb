```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Filtrando DataSet**

```python
# leitura de arquivo
df_aluno = pd.read_csv(
    "alunos-ingressantes-na-graduacao-pda-2017-2018.csv", encoding="ISO-8859-1"
)
df_2019 = pd.read_csv(
    "alunos-ingressantes-na-graduacao-pda-2020-2022.csv", encoding="ISO-8859-1"
)
df_2022 = pd.read_csv("ingressantes-2022-1.csv", encoding="ISO-8859-1")
df_2023 = pd.read_csv("ingressantes-2023-1.csv", encoding="ISO-8859-1")
df_2021 = pd.read_csv("alunos-ingressantes-graduacao-2021.csv", encoding="ISO-8859-1")

df_aluno2017 = pd.read_csv(
    "alunos-concluidos-da-graduacao-pda-2017-2018.csv", encoding="ISO-8859-1"
)
df_aluno2019 = pd.read_csv(
    "alunos-concluidos-da-graduacao-pda-2020-2022.csv", encoding="ISO-8859-1"
)
df_aluno2021 = pd.read_csv(
    "alunos-concluidos-graduacao-2021.csv", encoding="ISO-8859-1"
)
df_aluno2022 = pd.read_csv("concluintes-2022-1.csv", encoding="ISO-8859-1")
df_aluno2022_2 = pd.read_csv("concluintes-2022-2.csv", encoding="ISO-8859-1")
df_aluno2023 = pd.read_csv("concluintes-2023-1.csv", encoding="ISO-8859-1")
```

Ingressantes

```python
df_aluno = df_aluno[df_aluno["campus"] == "Itajuba"]

# 2017
df_2017 = df_aluno[df_aluno["ano_ingresso"] == 2017]
df_2017_sin = df_2017[df_2017["curso"] == "Sistemas de Informacao"]
df_2017_cco = df_2017[df_2017["curso"] == "Ciencias da Computacao"]

# 2018
df_2018 = df_aluno[df_aluno["ano_ingresso"] == 2018]
df_2018_sin = df_2018[df_2018["curso"] == "Sistemas de Informacao"]
df_2018_cco = df_2018[df_2018["curso"] == "Ciencias da Computacao"]

# Aqui so tem info do ano de 2019
df_2019 = df_2019[df_2019["campus"] == "Itajuba"]
df_2019_sin = df_2019[df_2019["curso"] == "Sistemas de Informacao"]
df_2019_cco = df_2019[df_2019["curso"] == "Ciencias da Computacao"]

# Aqui é so do ano de 2021 primeiro semestre
# df_2021 = df_2021[df_2021["campus"] == "Itajuba"]
df_2021_sin = df_2021[df_2021["curso"] == "Sistemas de Informacao"]
df_2021_cco = df_2021[df_2021["curso"] == "Ciencia da Computacao"]

# Aqui é so do ano de 2022 primeiro semestre
# df_2022 = df_2022[df_2022["campus"] == "Itajuba"]
df_2022_sin = df_2022[df_2022["curso"] == "Sistemas de Informacao"]
df_2022_cco = df_2022[df_2022["curso"] == "Ciencias da Computacao"]

# Aqui é so do ano de 2023 primeiro semestre
df_2023 = df_2023[df_2023["campus"] == "Itajuba"]
df_2023_sin = df_2023[df_2023["curso"] == "Sistemas de Informacao"]
df_2023_cco = df_2023[df_2023["curso"] == "Ciencias da Computacao"]

df_ingressante_sin = pd.concat(
    [df_2017_sin, df_2018_sin, df_2019_sin, df_2021_sin, df_2022_sin, df_2023_sin]
)
df_ingressante_cco = pd.concat(
    [df_2017_cco, df_2018_cco, df_2019_cco, df_2021_cco, df_2022_cco, df_2023_cco]
)

df_ingressante = pd.concat(
    [df_ingressante_sin, df_ingressante_cco]).reset_index()

df_ingressante
```

Concluintes

```python
# 2017
df_aluno2017 = df_aluno2017[df_aluno2017["campus"] == "Itajuba"]
df_2017 = df_aluno2017[df_aluno2017["ano_colacao"] == 2017]
df_2017_sin = df_2017[df_2017["curso"] == "Sistemas de informacao"]
df_2017_cco = df_2017[df_2017["curso"] == "Ciencia da Computacao"]

# 2018
df_2018 = df_aluno2017[df_aluno2017["ano_colacao"] == 2018]
df_2018_sin = df_2018[df_2018["curso"] == "Sistemas de informacao"]
df_2018_cco = df_2018[df_2018["curso"] == "Ciencia da Computacao"]

# Aqui so tem info do ano de 2019
df_aluno2019 = df_aluno2019[df_aluno2019["campus"] == 'Itajuba']
df_aluno2019_sin = df_aluno2019[df_aluno2019["curso"]
                                == "Sistemas de informacao"]
df_aluno2019_cco = df_aluno2019[df_aluno2019["curso"]
                                == "Ciencia da Computacao"]
```

```python
# Aqui so tem info do ano de 2021
df_aluno2021 = df_aluno2021[df_aluno2021["campus"] == "Itajuba"]
df_aluno2021_sin = df_aluno2021[df_aluno2021["curso"] == "Sistemas de Informacao"]
df_aluno2021_cco = df_aluno2021[df_aluno2021["curso"] == "Ciencia da Computacao"]
# CCO n tem valor algum em 2021

# Aqui é so do ano de 2022 primeiro semestre
df_aluno2022 = df_aluno2022[df_aluno2022["campus"] == "Itajuba"]
df_2022_sin = df_aluno2022[df_aluno2022["curso"] == "Sistemas de Informacao"]
df_2022_cco = df_aluno2022[df_aluno2022["curso"] == "Ciencia da Computacao"]


# Aqui é so do ano de 2022 segundosemestre
df_aluno2022_2 = df_aluno2022_2[df_aluno2022_2["campus"] == "Itajuba"]
df_2022_2_sin = df_aluno2022_2[df_aluno2022_2["curso"] == "Sistemas de Informacao"]
df_2022_2_cco = df_aluno2022_2[df_aluno2022_2["curso"] == "Ciencia da Computacao"]

df_2022_sin = pd.concat([df_2022_sin, df_2022_2_sin], verify_integrity=True)
df_2022_cco = pd.concat([df_2022_cco, df_2022_2_cco], verify_integrity=True)


# df_2022_sin = df_2022_sin.drop_duplicates(subset=None, keep="first", inplace=False)
# df_2022_cco = df_2022_cco.drop_duplicates(subset=None, keep="first", inplace=False)
# Aqui é so do ano de 2023 primeiro semestre
df_aluno2023 = df_aluno2023[df_aluno2023["campus"] == "Itajuba"]
df_2023_sin = df_aluno2023[df_aluno2023["curso"] == "Sistemas de informacao"]
df_2023_cco = df_aluno2023[df_aluno2023["curso"] == "Ciencia da Computacao"]
```

```python
df_concluinte_sin = pd.concat(
    [
        df_2017_sin,
        df_2018_sin,
        df_aluno2019_sin,
        df_aluno2021_sin,
        df_2022_sin,
        df_2023_sin,
    ],
    verify_integrity=False,
)
df_concluinte_cco = pd.concat(
    [
        df_2017_cco,
        df_2018_cco,
        df_aluno2019_cco,
        df_aluno2021_cco,
        df_2022_cco,
        df_2023_cco,
    ],
    verify_integrity=False,
)
df_concluinte = pd.concat(
    [df_concluinte_sin, df_concluinte_cco], verify_integrity=False
).reset_index()
```

**Unificação dos Ingressantes e Concluintes**

```python

df_junto = pd.concat([df_ingressante, df_concluinte],
                     verify_integrity=False).reset_index()

df_junto = df_junto.drop(
    ['level_0',  'index', 'modalidade', 'forma_ingresso'], axis=1)

concluinte = df_junto[df_junto["ano_colacao"]
                      != "NaN"].groupby('ano_colacao').count()
ingressante = df_junto[df_junto["ano_ingresso"]
                       != "NaN"].groupby('ano_ingresso').count()


novo = concluinte.drop(['id_discente', 'ano_ingresso', 'campus'], axis=1)
novo2 = ingressante['campus']
novo_df = pd.concat([novo, novo2], axis=1)
novo_df = novo_df.drop(["Unnamed: 6"], axis=1)
novo_df.columns = ['Concluintes', 'Ingressantes']

plt.rcParams['figure.figsize'] = [8, 6]
sns.set_context("notebook", font_scale=1.25, rc={
                "lines.linewidth": 4, 'axes.titlesize': 16})
novo_df.plot(kind='bar', stacked=False)
plt.title('Quantidade de ingressantes e Concluintes na Unifei')
plt.ylabel('Quantidade de pessoas')
plt.xlabel('Ano')
plt.show()
```
